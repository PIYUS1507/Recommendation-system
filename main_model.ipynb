```python
df = df[['title','overview','genres','tagline','vote_average','popularity']]
```

```python
df = df.drop_duplicates().reset_index(drop = True)
```

```python
df.isnull().sum()
```

```python
df['overview']= df['overview'].fillna(" ")
```

```python
df['tagline']= df['tagline'].fillna(" ")
```

```python
df= df.dropna(subset=['title'])
```

```python
df 
```

```python
df['genres'].iloc[0]
```

```python
import ast 

df['genres'] = df['genres'].apply(lambda x:" ".join(i['name'] for i in ast.literal_eval(x)))
```

```python
df['tag'] = df['overview'] +" "+df['genres']+" "+df['tagline']
```

```python
df
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
# nltk.download('stopwords')
# nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmetizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    #lowercase
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    words = text.split()

    words = [word for word in words if word not in stop_words]

    words = [lemmetizer.lemmatize(word) for word in words]

    return " ".join(words)
```

```python
df['tag'] = df['tag'].apply(preprocess_text)
```

```python
df['tag'][0]
```

```python
df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index,index=[df['title']]).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tdidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tdidf.fit_transform(df['tag'])
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title,n=10):
    if title not in indices:
        return ['Movie is not found']
    idx = indices[title]
    sim_score = cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
    similar_mov = sim_score.argsort()[::-1][1:n+1]
    return df['title'].iloc[similar_mov]
```

```python
recommend('Avatar',5)
```
